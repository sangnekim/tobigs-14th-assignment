latitude, longitude는 float 이지만 object형태로 df에 저장되있다  
float으로 타입변화를 하는데 값들 중 '-'이 존재해서 실수형으로 변환되지 않는 행 존재한다  
해당 값들은 결측치이다

```python
#수정해야될 행을 찾기 위해 이상치 있는 데이터 인덱스 확인
latitude_error = []
longitude_error = []
for i in range(len(df)):
    if '-' in df['latitude_x'].iloc[i]:
        latitude_error.append(i)
    if '-' in df['longitude_x'].iloc[i]:
        longitude_error.append(i)
print("latitude null in index", latitude_error)
print("longitude null in index", longitude_error)
```

```python
#결측치 인지 확인
df.iloc[[15,37,39,92,110]][['latitude_x','longitude_x']]
```

해당 인원은 latitude, longitude 정보가 없다

```python
#같은 지역 확인하기 위해 province와 city 칼럼을 합친 region 칼럼 생성
df['region'] = df['province_x']+'-'+df['city']
```

```python
# 이상치 처리
drop_list = []
for i in [15,37,39,92,110]:
    length = len(df[df.region == df.iloc[i]['region']])
    if length > 1: #같은 지역이 존재하면 나머지 값들을 이용해서 평균값을 할당
        class_index = df[df.region == df.iloc[i]['region']].index.tolist() # index_list
        sum_latitude = 0
        sum_longitude = 0
        for j in class_index:
            if j == i:
                continue
            else:
                sum_latitude += float(df['latitude_x'].iloc[j])
                sum_longitude += float(df['longitude_x'].iloc[j])
        mean_latitude = sum_latitude/(length-1)
        mean_longitutde = sum_longitude/(length-1)
        #평균 위도,경도를 새로 할당
        df.iloc[i,6] = str(mean_latitude)
        df.iloc[i,7] = str(mean_longitutde)
    else: #해당 지역에 자신만 존재하는 경우 위도,경도를 알 수 없으므로 drop
        drop_list.append(i) #하나씩 drop할 경우 index가 꼬일 수 있으므로 한번에 처리
df.drop(drop_list, inplace=True)
```

```python
max_confirmed = df['confirmed'].max() #신전치 사례
print(df.query('confirmed == 4511')[['latitude_x','longitude_x']])
sin_latitude, sin_longitude_x = df.query('confirmed == 4511')[['latitude_x','longitude_x']].iloc[0]
sincheonji = (float(sin_latitude),float(sin_longitude_x))
```

```python
from haversine import haversine #위도,경도를 이용해 거리 계산해주는 package import
```

```python
#계산을 위해 실수형으로 변환
df['latitude_x'] = df['latitude_x'].astype('float')
df['longitude_x'] = df['longitude_x'].astype('float')
#거리 구하는 함수 정의
def get_distance(df):
    return haversine((df['latitude_x'],df['longitude_x']), sincheonji)
#apply이용해서 새로운 컬럼 할당
df['distance(km)'] = df.apply(get_distance, axis = 1 )
```

```python
sns.scatterplot(data = df.query('confirmed < 4511'),
           x = 'distance(km)',
           y = 'confirmed')
plt.show()
```

```python
#'infection case'에 사용된 단어 리스트 생성
case_list = df['infection_case'].unique().tolist()
case_word_list = list(" ".join(case_list).split(' '))
```

```python
from collections import Counter #package import
cnt = Counter(case_word_list)
print(cnt.most_common(10)) #'infection case'에서 가장 많이 사용된 단어 top10
common_case_list = list(cnt.most_common(10))
#top3 단어 추출(교회, 병원, 센터)
top3_case_list = [common_case_list[i][0] for i in range(3)] 
```

```python
#(교회, 병원, 센터, 기타)로 다시 한번 케이스 분류
def find_case(df):
    infection_case = df['infection_case']
    temp_case = 'etc' #기타
    for place in top3_case_list:
        if place in infection_case:
            temp_case = place
    return temp_case
df['case'] = df.apply(find_case, axis=1) #'case'컬럼에 새로운 케이스 할당
```

```python
sns.countplot(df['case'])
plt.show()
```

'infection_case'를 문자열 처리를 통해서 다시 한번 케이스 분류를 통해 대표적인 기관들(교회, 센터, 병원)별 확진사례를 확인할 수 있다  
분류되지 않은 etc의 경우를 제외하면, 교회와 병원에서 감염사례가 발생한것을 확인할 수 있다
